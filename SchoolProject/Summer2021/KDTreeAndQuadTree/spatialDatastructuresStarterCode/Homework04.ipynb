```python
# imports and setup
import seaborn as sns
import statsmodels.formula.api as sm
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm

import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
```

```python
csvK = pd.read_csv("KDiff")
csvN = pd.read_csv("NDiff")
csvD = pd.read_csv("DDiff")

# print(csvN.head())
# print(csvK.head())
# print(csvD.head())

sns.lineplot(x="kvalue", y="Time", data=csvK, hue="SortMethod")
KData = sm.ols(formula="Time ~ kvalue + SortMethod", data=csvK).fit()
KData.summary()
```

```python
sns.lineplot(x="Nvalue", y="Time", data=csvN, hue="SortMethod")
NData = sm.ols(formula="Time ~ Nvalue + SortMethod", data=csvN).fit()
NData.summary()
```

```python
sns.lineplot(x="Dim", y="Time", data=csvD, hue="SortMethod")
DData = sm.ols(formula="Time ~ Dim + SortMethod", data=csvD).fit()
DData.summary()
```
